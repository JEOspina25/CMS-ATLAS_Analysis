# Z Analysis

```python
def Z(S,B): return S/np.sqrt(S+B)
def w(L,cs,N_mc): return (L*cs)/N_mc
```

#### Signal S

```python
cs_s = 0.2 * 1000 # fb (cross section generacion MG) ** 1 pb -> 1000 fb 
N_mc_s = 50000 # numero eventos inicial
N_s1 = len(data_signal_cut1) # numero de eventos final corte 1
N_s2 = len(data_signal_cut2) # numero de eventos final corte 2
L = 3000 # fb^-1
```

```python
S_1 = N_s1 * w(L,cs_s,N_mc_s)
S_2 = N_s2 * w(L,cs_s,N_mc_s)
print(S_1, " -- ", S_2)
```

#### Signal B1

```python
cs_bbbar = 6.173970e+03 * 1000 # fb (cross section generacion MG) ** 1 pb -> 1000 fb 
N_mc_bbbar = 848568 # numero eventos inicial
N_bbbar1 = len(data_bbbar_cut1) # numero de eventos final corte 1
N_bbbar2 = len(data_bbbar_cut2) # numero de eventos final corte 2
L = 3000 # fb^-1
```

```python
B1_1 = N_bbbar1 * w(L,cs_bbbar,N_mc_bbbar)
B1_2 = N_bbbar2 * w(L,cs_bbbar,N_mc_bbbar)
print(B1_1, " -- ", B1_2)
```

#### Signal B2

```python
cs_qcd = 2.836e+06 * 1000 # fb (cross section generacion MG) ** 1 pb -> 1000 fb 
N_mc_qcd = 1000000 # numero eventos inicial
N_qcd1 = len(data_QCD_cut1) # numero de eventos final corte 1
N_qcd2 = len(data_QCD_cut1) # numero de eventos final corte 2
L = 3000 # fb^-1
```

```python
B2_1 = N_qcd1 * w(L,cs_qcd,N_mc_qcd)
B2_2 = N_qcd2 * w(L,cs_qcd,N_mc_qcd)
print(B2_1, " -- ", B2_2)
```

#### B1 + B2

```python
B_1 = B1_1 + B1_2
B_2 = B2_1 + B2_2
```

# Z

```python
print("Z_A =",Z(S_1,B_1))
print("Z_B =",Z(S_2,B_2))
```
